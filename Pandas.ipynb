```python
import pandas as pd 

a = pd.Series({'a':9 ,'b':2 ,'c':1 ,'d':33})
q = pd.Series({'a':19,'b':12,'c':11,'d':3})
c = pd.Series({'a':29,'b':22,'c':12,'d':23})
v = pd.Series({'a':5, 'b':4, 'c':8, 'd':93})

grades=pd.DataFrame({'math':a,'physics':q, 'german':c, 'biologie':v})

grades['Total']=(grades['math'] + grades['physics'] )/2 
print(grades.query('Total >14'))
```

```python
import pandas as pd 
def make_df(cols,ind):
    data={ c:[str(c)+ str(i) for i in ind] for c in cols}
    
    return pd.DataFrame(data,ind)

print(make_df('ABC', range(3)))
    
    
    
```
